# **Step 2: Data Preprocessing**
**Preprocess bioactivity data (IC50) for the selected Acetylcholinesterase target.**


**Inspect Missing Values**


```python
# =============================================================================
# CELL 27 - CODE: Check missing values
# =============================================================================
print("Missing values in key columns:")
print(f"standard_type: {df1['standard_type'].isna().sum()}")
print(f"standard_value: {df1['standard_value'].isna().sum()}")
print(f"canonical_smiles: {df1['canonical_smiles'].isna().sum()}")
```

**Filter Rows with Valid Bioactivity Values**


```python
# =============================================================================
# CELL 29 - CODE: Filter valid values
# =============================================================================
df2 = df1[df1["standard_value"].notna()]
print(f"Records after removing missing values: {len(df2)}")
df2.head()
```

**Standardize Units - Keep only nM (nanomolar)**


```python
# =============================================================================
# CELL 31 - CODE: Check and filter units
# =============================================================================
print("Unit distribution:")
print(df2['standard_units'].value_counts())

# Keep only nM units
df2 = df2[df2['standard_units'] == 'nM']
print(f"\nRecords after keeping only nM: {len(df2)}")
```

**Assign Bioactivity Classes**
Define active, intermediate, and inactive classes based on IC50 values.

- **Active:** IC50 ≤ 1000 nM
- **Intermediate:** 1000 < IC50 < 10000 nM  
- **Inactive:** IC50 ≥ 10000 nM


```python
# =============================================================================
# CELL 33 - CODE: Assign bioactivity classes
# =============================================================================
bioactivity_class = []
for value in df2.standard_value:
    value = float(value)
    if value >= 10000:
        bioactivity_class.append("inactive")
    elif value <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")

print(f"Total compounds classified: {len(bioactivity_class)}")
```

**Extract Relevant Columns**


```python
# =============================================================================
# CELL 35 - CODE: Extract columns
# =============================================================================
molecule_ids = df2.molecule_chembl_id.tolist()
canonical_smiles = df2.canonical_smiles.tolist()
standard_values = df2.standard_value.tolist()
```

```python
# =============================================================================
# CELL 36 - CODE: Create data tuple
# =============================================================================
data = list(zip(
    molecule_ids,
    canonical_smiles,
    standard_values,
    bioactivity_class,
))
```

**Create Preprocessed Bioactivity Dataset**


```python
# =============================================================================
# CELL 38 - CODE: Create final dataframe
# =============================================================================
df3 = pd.DataFrame(
    data,
    columns=[
        "molecule_chembl_id",
        "canonical_smiles",
        "standard_value",
        "bioactivity_class",
    ]
)
df3.head()
```

**Remove Compounds without Valid SMILES**
Drop rows with NaN, empty, or None SMILES values.


```python
# =============================================================================
# CELL 40 - CODE: Clean SMILES
# =============================================================================
print(f"Records before SMILES cleaning: {len(df3)}")

df3 = df3.dropna(subset=["canonical_smiles"])
df3 = df3[df3["canonical_smiles"].str.lower() != "none"]
df3 = df3[df3["canonical_smiles"].str.strip() != ""]

print(f"Records after SMILES cleaning: {len(df3)}")
df3.head()
```

**Remove Duplicate Molecules**


```python
# =============================================================================
# CELL 42 - CODE: Remove duplicates
# =============================================================================
print(f"Records before removing duplicates: {len(df3)}")
df3 = df3.drop_duplicates(subset=['canonical_smiles'])
print(f"Records after removing duplicates: {len(df3)}")
```

**Dataset Summary**


```python
# =============================================================================
# CELL 44 - CODE: Summary statistics
# =============================================================================
print("=" * 50)
print("FINAL DATASET SUMMARY")
print("=" * 50)
print(f"Target: Acetylcholinesterase (CHEMBL220)")
print(f"Total curated records: {len(df3)}")
print(f"\nBioactivity class distribution:")
print(df3['bioactivity_class'].value_counts())
```
